# 2.3: 单比特量子态的测量与期望

测量：
* 经典测量：例如一堆硬币，我们通过直接"看"硬币即可知道它所处的状态，且"看"这个动作并不会改变其状态。
* 量子测量：通过测量操作实现，由量子算符刻画，量子态作用于测量算符后（被测量后）会使量子态发生坍缩，量子态发生改变。

核心基础，量子态的波恩概率诠释：

$p(s) := |\varphi_s|^2 = \varphi_s^*\varphi_s = \langle\varphi|s\rangle\langle s|\varphi\rangle = |\langle\varphi|s\rangle|^2$

> 其中 $|s\rangle\langle s|$ 是投影算符


## 例：投影测量 (projective measurement)

* 利用泡利算符$\hat{\sigma}^{(Z)}$的本征态进行测量，相应的测量算符为本征态构成的投影算符

    $\hat{M}^{(0)} = |0\rangle\langle0|,\quad \hat{M}^{(1)} = |1\rangle\langle1|$
* 考虑量子态$|+\rangle$，根据波恩诠释，对应的概率为相应测量算符的期望值，满足

    $p_0 = \langle+|\hat{M}^{(0)}|+\rangle = 0.5$

    $p_1 = \langle+|\hat{M}^{(1)}|+\rangle = 0.5$


## 测量算符的完备性条件（概率归一化条件）：
测量的基底不一定需要正交归一，任意完备基底$\{|s\rangle\}$都可以用来进行测量，“完备”指满足

$\sum_m \hat{M}^{(m)} = \hat{I} \Leftrightarrow \sum_m p_m = 1$

> 这种测量叫做 POVM（positive operator-valued measure），即正算符值测量

> POVM 测量和投影测量的区别在于，投影测量需要正交归一的基底，而 POVM 测量只需要完备的基底


## 测量后的量子态

测量后的量子态为测量算符作用后的态

$|\varphi^{(s)}\rangle = \frac{1}{Z}\hat{M}^{(s)}|\varphi\rangle$

其中Z为归一化系数。

> 公式解读：
> * $|\varphi^{(s)}\rangle$为测量后量子态，被置为 $|s\rangle$
> * $\hat{M}^{(s)}$为取得量子态 $|s\rangle$ 的测量算符，$|\varphi\rangle$为测量前量子态
> * 这个公式成立是量子计算的基本假设

* 在真实的量子计算机中，测量后设备给出的结果为"经典数据" $s$, 例如量子比特的测量结果为一串经典比特(0或1的二进制数)，而不是直接返回概率值 $p_s$ 或测量后的量子态$|\varphi^{(s)}\rangle$。
    * 测量也称量子态采样 quantum state sampling
    * 量子采样所得的每个经典比特称为一个样本或者构型

* 概率值可由测量结果估计出来，满足

    $p_s \approx \frac{N_s}{\sum_s N_s}$

    其中$N_s$为比特串中s出现的次数。


## 对比
|| 测量前的量子态 | 测量算符（基底） | 采样构型 | 测量后的量子态 | 获得各构型的概率 |
|---|---|---|---|---|---|
|一般形式| $\|\varphi\rangle$ | $\{\hat{M}^{(s)} = \|s\rangle\langle s\|\}$ | $s$ | $\|\varphi^{(s)}\rangle = \frac{1}{Z}\hat{M}^{(s)}\|\varphi\rangle$ | $p_s = \langle\varphi\|\hat{M}^{(s)}\|\varphi\rangle$ |
|Qubit 与测量投影| $\|\varphi\rangle$ | $\|0\rangle\langle0\|$ 与 $\|1\rangle\langle1\|$ | 0或1 | $\|0\rangle$ 或 $\|1\rangle$ | $\langle\varphi\|0\rangle\langle0\|\varphi\rangle$ 或 $\langle\varphi\|1\rangle\langle1\|\varphi\rangle$ |

## 量子态层析 Quantum State Tomography

通过一组完备的投影测量，只能估计出测量结果的概率分布，但并不能获得量子态的全部信息，特别是相位信息。

> 例：$|+'\rangle = \frac{\sqrt{2}}{2}|0\rangle + \frac{\sqrt{2}}{2}e^{i\alpha}|1\rangle$ 与 $|+\rangle = \frac{\sqrt{2}}{2}|0\rangle + \frac{\sqrt{2}}{2}|1\rangle$
>
> 二者在 $\hat{M}^{(0)} = |0\rangle\langle0|$ 与 $\hat{M}^{(1)} = |1\rangle\langle1|$ 这组测量基底下，不同的 $\alpha$ 给出完全等同的概率分布，有 $p_0 = p_1 = 0.5$，故无法通过该组测量获得 $\alpha$ 的取值。
>
> 但在测量算符 $\hat{M}^{(+)} = |+\rangle\langle+|$ 与 $\hat{M}^{(-)} = |-\rangle\langle-|$ 下，有
>  * $p_+ = \langle+'|\hat{M}^{(+)}|+'\rangle = 0.5(1 + \cos \alpha)$
>  * $p_- = \langle+'|\hat{M}^{(-)}|+'\rangle = 0.5(1 - \cos \alpha)$

要完全获得量子态系数，需要进行多组测量。通过测量结果反推量子态系数的方法被称为量子态层析 (quantum state tomography)。

全局相因子 (global phase factor) 是不能通过测量结果推断出来的：

考虑 $|\varphi'\rangle = e^{i\alpha}|\varphi\rangle$，可得 $|\varphi'\rangle$ 关于任意量子算符（当然包括测量算符）的期望值与 $\alpha$ 无关

$\langle\varphi'|\hat{O}|\varphi'\rangle = \langle\varphi|e^{-i\alpha}\hat{O}e^{i\alpha}|\varphi\rangle = \langle\varphi|\hat{O}|\varphi\rangle$

> 注：$|\varphi'\rangle^\dagger = (e^{i\alpha}|\varphi\rangle)^\dagger = e^{-i\alpha}\langle\varphi|$

全局相因子可看作是一种"规范自由度" (gauge degrees of freedom)，即为不影响物理观测结果的数学上的自由度。

## 从算符的本征值分解理解量子期望

期望值为本征值 $\lambda_s$ 在概率分布 $p_s = |\langle u^{(s)}|\psi\rangle|^2$ 下的加权平均

$\langle \hat{O} \rangle \equiv \langle\psi|\hat{O}|\psi\rangle = \langle\lambda_s\rangle$

概率 $p_s$ 为 $|\psi\rangle$ 在 $\hat{O}$ 的本征态为测量基底下的测量概率。

> 复习：
> * 算符的本征值分解 $\hat{O} = \sum_s \lambda_s |u^{(s)}\rangle\langle u^{(s)}|$
>
> * 算符的本征方程 $\hat{O}|u^{(s)}\rangle = \lambda_s|u^{(s)}\rangle$

> 证明：
> $$
> \begin{align*}
> \langle \hat{O} \rangle &\equiv \langle \psi|\hat{O}|\psi \rangle \\
> &= \langle \psi|\sum_s \lambda_s |u^{(s)} \rangle \langle u^{(s)}|\psi \rangle \\
> &= \sum_s \lambda_s |\langle u^{(s)}|\psi \rangle|^2 \\
> &= \sum_s \lambda_s |\langle u^{(s)}|\psi \rangle|^2 \\
> &= \sum_s \lambda_s p_s \\
> &= \langle \lambda_s \rangle
> \end{align*}
> $$
