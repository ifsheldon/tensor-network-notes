# 单比特量子算符

量子算符 (quantum operator)：又被称为量子门 (quantum gate)，可理解为对量子态的操作

算符以 $|s'\rangle\langle s|$ 为基底，其一般形式为

$\hat{O} = \sum_{s's} O_{s's}|s'\rangle\langle s|$

其中，右矢与左矢外积运算的一般定义为

$\hat{O} = |\psi\rangle\langle\varphi| = \sum_{s's} \psi_{s'}\varphi_s^*|s'\rangle\langle s|$

所得系数刚好为量子态系数 $\psi^†$ 与 $\varphi$ 的外积，即 $O = \psi^†\otimes\varphi$

> 例：$|0\rangle\langle1|$的系数满足
> $\begin{bmatrix}1\\0\end{bmatrix}[0\quad1]=\begin{bmatrix}0&1\\0&0\end{bmatrix}$

* 量子算符作用于量子态的运算可化为矩阵与向量的乘积
    
    考虑$\hat{O} = \sum_{s's}O_{s's}|s'\rangle\langle s|$，$|\varphi\rangle = \sum_{s=0,1}\varphi_s|s\rangle$，有 $|\varphi'\rangle = \hat{O}|\varphi\rangle = \sum_{s's}O_{s's}\varphi_s|s'\rangle$

    即 $\varphi' = O\varphi$

    > 例：$|\varphi\rangle = \frac{\sqrt{2}}{2}|0\rangle + \frac{\sqrt{2}}{2}|1\rangle$，$\hat{O} = |0\rangle\langle0| - |1\rangle\langle1|$，则
    >
    > $|\varphi'\rangle = \hat{O}|\varphi\rangle = \frac{\sqrt{2}}{2}|0\rangle - \frac{\sqrt{2}}{2}|1\rangle$
    > 
    > 其中 $O\varphi = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}\begin{bmatrix} \frac{\sqrt{2}}{2} \\ \frac{\sqrt{2}}{2} \end{bmatrix} = \begin{bmatrix} \frac{\sqrt{2}}{2} \\ -\frac{\sqrt{2}}{2} \end{bmatrix}$

* 量子算符作用于量子算符的运算可化为矩阵与矩阵的乘积
    
    考虑$\hat{X} = \sum_{s's}X_{s's}|s'\rangle\langle s|$，$\hat{Y} = \sum_{s's}Y_{s's}|s'\rangle\langle s|$，有 $\hat{O} = \hat{X}\hat{Y} = \sum_{s''s's}X_{s''s'}Y_{s's}|s''\rangle\langle s|$

    即 $O = XY$

* 量子算符$\hat{O}$关于量子态$|\varphi\rangle$的量子期望（或观测值）定义为

    $\langle\hat{O}\rangle = \langle\varphi|\hat{O}|\varphi\rangle$

    * 给定基底下，有 $\langle\hat{O}\rangle = \varphi^\dagger O\varphi$
    * 例：对于一个量子态$|\psi\rangle$ 我们可以定义一个投影算符 (projector) $\hat{P} := |\psi\rangle\langle\psi|$
        * 性质1：其系数矩阵为 $|\psi\rangle$ 系数（向量）与其转置共轭的外积，即 $|\psi\rangle\langle\psi| = \sum_{s's}\psi^*_{s'}\psi_s|s'\rangle\langle s|$
        * 性质2：$|\psi\rangle$ 构成的投影算符将任意量子态 $|\varphi\rangle$ 投影到 $|\psi\rangle$ 的方向，即

            $\hat{P}|\varphi\rangle = (|\psi\rangle\langle\psi|)|\varphi\rangle = |\psi\rangle\langle\psi|\varphi\rangle \sim |\psi\rangle$ 
        * 性质3：量子态 $|\varphi\rangle$ 关于 $|\psi\rangle$ 的投影算符的期望值给出 $|\varphi\rangle$ 中测量获得 $|\psi\rangle$ 态的概率

            $p(|\psi\rangle) = \langle\varphi|\psi\rangle\langle\psi|\varphi\rangle = |\langle\varphi|\psi\rangle|^2$
            > 复习：系数的模方描述该量子系统处于对应（某个基底的）状态的概率 $p(s) := |\varphi_s|^2 = \varphi_s^*\varphi_s = \langle\varphi|s\rangle\langle s|\varphi\rangle = |\langle\varphi|s\rangle|^2$

* 量子算符的本征值分解

    $\hat{O} = \sum_s \lambda_s |u^{(s)}\rangle\langle u^{(s)}|$

    其中$|u^{(s)}\rangle$为$\hat{O}$的本征态，$\lambda_s$为本征值，满足本征方程 $\hat{O}|u^{(s)}\rangle = \lambda_s|u^{(s)}\rangle$

    > 矩阵的本征值分解也可以写为 $O = \sum_s \lambda_s u^{(s)} * u^{(s)\dagger}$, $u^{(s)}$ 是第 s 个本征向量

    性质：

    1. $|u^{(s)}\rangle$ 的系数 $u^{(s)}$ 为 $\bar{O}$ 的系数矩阵 $O$ 的本征向量，满足： $Ou^{(s)} = \lambda_s u^{(s)}$

    2. 设 $O$ 的本征值分解为 $O = U\lambda U^\dagger$，则有 $U_{s's} = \langle s'|u^{(s)}\rangle$

## 核心：量子态的表示与运算 ↔ 张量的表示与运算

对于单比特量子态：
- 量子态 ↔ 二维归一化向量
- 量子算符 ↔ $(2\times2)$矩阵
- 量子态间的内积、外积 ↔ 向量的内积、外积
- 量子算符作用于量子态 ↔ 矩阵与向量的乘法运算
- 量子算符作用于量子算符 ↔ 矩阵乘法
- 量子算符的本征值分解 ↔ 系数矩阵的本征值分解

## 热态


热态 $\hat{\rho}$ 本质上为 Hermitian 算符，满足如下约束条件
$$\text{Tr}(\hat{\rho}) = 1, \text{Tr}(\hat{\rho}^2) \leq 1$$

当 $\text{Tr}(\hat{\rho}^2) = 1$ 时，有 $\hat{\rho}=|\varphi\rangle\langle\varphi|$，称其为**纯态 (pure state)**，系数矩阵为单秩矩阵（见1.8节）；$\text{Tr}(\hat{\rho}^2) < 1$ 时热态不能写成外积形式，称其为**混合态 (mixed state)**，系数矩阵不为单秩矩阵。

由热态 $\hat{\rho}$ 获得某个态（正交归一基的基态） $|s\rangle$ 的概率满足
$$p_s = \langle s|\hat{\rho}|s\rangle$$

因此，$\text{Tr}(\hat{\rho}) = \sum_s\langle s|\hat{\rho}|s\rangle = \sum_s p_s = 1$ 对应于概率归一化

> $\text{Tr}(\hat{\rho}) = \sum_s\langle s|\hat{\rho}|s\rangle$ 的等号成立是因为：
> 1. Trace 是矩阵对角线的和，取对角线的和等价于用最简单的 one-hot 基执行等式右边的式子
> 2. Trace 是基底无关的，所以特定矩阵（选择任意正交归一基）的 Trace 都不变，都等于 one-hot 基求出来的 trace，也就是矩阵对角线的和
> 
> Trace 基底无关性证明：
> 1. **基变换对应相似变换**  
>     设两组正交归一基 $\{|s\rangle\}$ 和 $\{|e_k\rangle\}$，通过幺正矩阵 $U$ 联系：  
>     $$|e_k\rangle = \sum_s U_{ks} |s\rangle, \quad U^\dagger U = I.$$  
>     算符 $\hat{\rho}$ 在基 $\{|e_k\rangle\}$ 下的矩阵为：  
>     $$\rho' = U^\dagger \rho U,$$  
>     其中 $\rho$ 是原基下的矩阵。  
>
> 2. **迹在相似变换下不变**  
>     对任意矩阵 $\rho$ 和幺正矩阵 $U$，有恒等式：  
>     $$
>     \text{Tr}(\rho') = \text{Tr}(U^\dagger \rho U) = \text{Tr}(\rho U U^\dagger) = \text{Tr}(\rho).
>     $$  
>     因此，无论基如何选取，$\text{Tr}(\hat{\rho}) = \sum_s \langle s|\hat{\rho}|s\rangle$ 的值始终相同。  


热态 $\hat{\rho}$ 关于算符 $\hat{O}$ 的量子期望满足 $$\langle O\rangle = \text{Tr}(\hat{\rho}\hat{O})$$
